```python
import networkx as nx # untuk membuat grafik
import matplotlib.pyplot as plt # untuk plot grafik
```

```python
def show_graph(G, pos):# fungsi untuk menampilkan graf
    nx.draw(G, pos, with_labels=True,# menggambar graf dengan label node
            node_size=2000,
            node_color='blue',
            font_size=12)
    plt.show()# menampilkan graf ke layar

G = nx.Graph()# membuat graf kosong (tidak berarah)

edges = [# membuat hubungan antar node (edges)
    ('A', 'B'), ('A', 'C'),
    ('B', 'D'), ('B', 'E'),
    ('C', 'F'), ('C', 'G')
]
G.add_edges_from(edges)# menambahkan semua edge ke graf

pos = {# menentukan posisi tiap node (x, y)
    'A': (0, 2),
    'B': (-1, 1),
    'C': (1, 1),
    'D': (-1.5, 0),
    'E': (-0.5, 0),
    'F': (0.5, 0),
    'G': (1.5, 0)
}

show_graph(G, pos)# menampilkan graf sesuai posisi
# DFS dengan depth limit = 2
dfs_limit = list(nx.dfs_edges(G, source='A', depth_limit=2)) #DFS dengan batas

print("\nDFS depth limit = 2:") 
i = 1
for u,v in dfs_limit: #loop edge
    print(f"{i}. {u} - {v}") #tampilin edge
    i += 1
```
