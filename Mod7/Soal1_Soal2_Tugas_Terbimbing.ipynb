```python
import networkx as nx
import matplotlib.pyplot as plt
```

SOAL 1

```python
def show_graph(G, pos=None, title=''):
    nx.draw(
        G, pos,
        with_labels=True,
        node_color="#4D26B5",
        node_size=1000,
        font_color="white",
        font_weight="bold",
        font_size= "xx-small",
        width=2,
        edge_color="black"
    )

    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(
        G, pos, 
        edge_labels=edge_labels, 
        font_color='red',
        font_size= 6
    )

    plt.title(title)
    plt.show()

pos = {
    'Arad': (0, 2),
    'Zerind': (-1, 3),
    'Oradea': (-1, 4),
    'Timisoara': (0, 0),
    'Lugoj': (1, -1),
    'Mehadia': (1, -2),
    'Drobeta': (1, -3),
    'Craiova': (3, -3),
    'Rimnicu Vilcea': (3, 0),
    'Sibiu': (2, 2),
    'Fagaras': (4, 2),
    'Pitesti': (4, -1),
    'Bucharest': (6, -1),
    'Giurgiu': (6, -3),
    'Urziceni': (7, 0),
    'Hirsova': (9, 0),
    'Eforie': (10, -1),
    'Vaslui': (8, 2),
    'Iasi': (7, 3),
    'Neamt': (6, 4)
}

edges = [
    ('Arad','Zerind',75), ('Zerind','Oradea',71),
    ('Arad','Sibiu',140), ('Oradea','Sibiu',151),
    ('Arad','Timisoara',118),
    ('Timisoara','Lugoj',111), ('Lugoj','Mehadia',70),
    ('Mehadia','Drobeta',75), ('Drobeta','Craiova',120),
    ('Craiova','Pitesti',138), ('Craiova','Rimnicu Vilcea',146),
    ('Rimnicu Vilcea','Sibiu',80), ('Rimnicu Vilcea','Pitesti',97),
    ('Sibiu','Fagaras',99), ('Fagaras','Bucharest',211),
    ('Pitesti','Bucharest',101),
    ('Bucharest','Giurgiu',90),
    ('Bucharest','Urziceni',85),
    ('Urziceni','Hirsova',98), ('Hirsova','Eforie',86),
    ('Urziceni','Vaslui',142), ('Vaslui','Iasi',92),
    ('Iasi','Neamt',87)
]

G = nx.Graph()
G.add_weighted_edges_from(edges)

show_graph(G, pos, "Graf Eropa")

#penjelasan singkat
path = nx.astar_path(G, 'Arad', 'Bucharest',    # mencari jalur dari Arad ke Bucharest (UCS karena heuristic = 0)
                heuristic=lambda a,b: 0,    # tidak menggunakan perkiraan jarak → jadi Uniform Cost Search
                weight='weight')            # menggunakan nilai weight sebagai biaya/jarak

cost = nx.path_weight(G, path, weight='weight') # menghitung total cost dari jalur yang ditemukan

print("Path:", path)
print("Total Cost:", cost)
```

```python
def show_graph(G, pos=None, title=''):
    nx.draw(
        G, pos,
        with_labels=True,
        node_color="#4D26B5",
        node_size=2000,
        font_color="white",
        font_weight="bold",
        font_size= "8",
        width=3,
        edge_color="black"
    )

    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(
        G, pos, 
        edge_labels=edge_labels, 
        font_color='red',
        font_size= 11
    )

    plt.title(title)
    plt.show()

pos_jawa = {
    'Jakarta': (0, 2),
    'Bandung': (1, 1),
    'Cirebon': (2, 2),
    'Semarang': (4, 2),
    'Yogyakarta': (3, 0),
    'Surakarta': (4, 1),
    'Surabaya': (6, 2),
    'Malang': (6, 0)
}

edges_jawa = [
    ('Jakarta','Cirebon',327),
    ('Jakarta','Bandung',270),
    ('Bandung','Cirebon',120),
    ('Bandung','Yogyakarta',373),
    ('Cirebon','Semarang',305),
    ('Cirebon','Yogyakarta',210),
    ('Yogyakarta','Semarang',109),
    ('Yogyakarta', 'Surakarta',60),
    ('Semarang','Surakarta',97),
    ('Surakarta','Malang',370),
    ('Semarang','Surabaya',369),
    ('Surabaya','Malang',94)
]

G = nx.Graph()
G.add_weighted_edges_from(edges_jawa)

show_graph(G, pos_jawa, "Graf Pulau Jawa")

#penjelasan singkat
path2 = nx.astar_path(G, 'Bandung', 'Malang',    # mencari jalur dari Arad ke Bucharest (UCS karena heuristic = 0)
                heuristic=lambda a,b: 0,    # tidak menggunakan perkiraan jarak → jadi Uniform Cost Search
                weight='weight')            # menggunakan nilai weight sebagai biaya/jarak

cost2 = nx.path_weight(G, path2, weight='weight') # menghitung total cost dari jalur yang ditemukan

print("Path:", path2)
print("Total Cost:", cost2)
```
