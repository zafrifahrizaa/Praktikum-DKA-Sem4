```python
import networkx as nx
import matplotlib.pyplot as plt

# fungsi tampil graf
def show_graph(G, pos=None, title=''):
    nx.draw(G, pos,
            with_labels=True,
            node_color="blue",
            node_size=1000,
            font_color="white")
    plt.title(title)
    plt.show()

# membuat graf tree
G = nx.Graph()

# menambah edge (contoh struktur tree)
edges = [
    ('A','B'), ('A','C'),
    ('B','D'), ('B','E'),
    ('C','F'), ('C','G')
]

G.add_edges_from(edges)

# posisi node biar seperti tree
pos = {
    'A': (0,3),
    'B': (-1,2), 'C': (1,2),
    'D': (-1.5,1), 'E': (-0.5,1),
    'F': (0.5,1), 'G': (1.5,1)
}

# tampilkan graf
show_graph(G, pos, "Graf Tree")

# BFS dari node A
bfs_edges = list(nx.bfs_edges(G, source='A'))

# urutan BFS edge
print("Urutan BFS (edge):")
for i, (u,v) in enumerate(bfs_edges, 1):
    print(f"{i}. {u} -> {v}")

# urutan BFS node
bfs_nodes = ['A'] + [v for u,v in bfs_edges]

print("\nUrutan BFS (node):")
print(" -> ".join(bfs_nodes))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def show_graph(G, pos=None, title=''):
    labels = nx.get_edge_attributes(G,'weight')
    nx.draw(G, pos,
            with_labels=True,
            node_color="green",
            node_size=3000,
            font_size= 10,
            font_color="white")
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.title(title)
    plt.show()

# buat graf berbobot
G = nx.Graph()

edges = [
    ('Bandung','Jakarta',150),
    ('Bandung','Yogyakarta',380),
    ('Jakarta','Semarang',450),
    ('Yogyakarta','Surabaya',330),
    ('Semarang','Surabaya',350)
]

G.add_weighted_edges_from(edges)

# posisi bebas
pos = nx.spring_layout(G)

# tampilkan graf
show_graph(G, pos, "Graf Kota Berbobot")

# BFS dari Bandung
bfs_edges = list(nx.bfs_edges(G, source='Bandung'))

print("Urutan BFS (edge):")
for i, (u,v) in enumerate(bfs_edges,1):
    print(f"{i}. {u} -> {v}")
```
