```python
import networkx as nx

G = nx.Graph()
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('C','D'), ('D','E')])

print("Degree:", dict(G.degree()))
print("Neighbor D:", list(G.neighbors('D')))

if nx.cycle_basis(G):
    print("Ada siklus")
else:
    print("Tidak ada siklus")
```

```python
import networkx as nx
# Membuat graf berbobot
G2 = nx.Graph()

G2.add_weighted_edges_from([
    ('A','B',4),
    ('A','C',2),
    ('B','C',1),
    ('B','D',5),
    ('C','D',8),
    ('C','E',10),
    ('D','E',2)
])

# Update weight C-E menjadi 1
G2['C']['E']['weight'] = 1
# Hitung ulang
new_path = nx.shortest_path(G2, source='A', target='E', weight='weight')
new_distance = nx.shortest_path_length(G2, source='A', target='E', weight='weight')

print("Jalur baru:", new_path)
print("Total weight baru:", new_distance)
```
